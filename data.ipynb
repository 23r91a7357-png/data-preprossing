```python
import pandas as pd
import numpy as np
data={'age':['25','27','np nan','100','30'],
      'salary':['50000','51000','50500','52000','51000'],
      'city':['A','A','B','A','A']}
df=pd.DataFrame(data)

# Convert age and salary to numeric, coercing errors will turn 'np nan' into actual NaN
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

print("original data")
print(df)

df['age']=df['age'].fillna(df['age'].mean())
print("\n after handling missing values:")
print(df)

Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
df=df[(df['age']>=lower_limit)&
      (df['age']<=upper_limit)]
print("\n after noise deduction (outlier removal)")
print(df)

Q1=df['salary'].quantile(0.25)
Q3=df['salary'].quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
df=df[(df['salary']>=lower_limit)&
      (df['salary']<=upper_limit)]
print("\n after noise deduction (outlier removal)")
print(df)

df=df.drop_duplicates()
print("\n after removing redundant (duplicates) records")
print(df)
```
